```python
# Se eliminan las URL's de la base de datos:
data2.drop(["listing_url","host_url","picture_url","host_thumbnail_url","host_picture_url","id"], axis = 1, inplace = True)
# Se eliminan las columnas con valores nulos (Del más del 80%) que tenemos en las bases del equipo
data2.drop(["neighbourhood_group_cleansed","calendar_updated","license"], axis = 1, inplace = True)
data2
```

## Limpieza

```python
# Se eliminan los símbolos que hacen que la columna de precio sea de tipo objeto

data2['host_response_rate'] = data2['host_response_rate'].str.replace('%', '')
data2['host_acceptance_rate'] = data2['host_acceptance_rate'].str.replace('%', '')
data2['host_is_superhost'] = data2['host_is_superhost'].str.replace('f', '0')
data2['host_is_superhost'] = data2['host_is_superhost'].str.replace('t', '1')
data2['host_has_profile_pic'] = data2['host_has_profile_pic'].str.replace('f', '0')
data2['host_has_profile_pic'] = data2['host_has_profile_pic'].str.replace('t', '1')
data2['host_identity_verified'] = data2['host_identity_verified'].str.replace('f', '0')
data2['host_identity_verified'] = data2['host_identity_verified'].str.replace('t', '1')
data2['instant_bookable'] = data2['instant_bookable'].str.replace('f', '0')
data2['instant_bookable'] = data2['instant_bookable'].str.replace('t', '1')

# Se convierte a numérico

data2['host_response_rate'] = pd.to_numeric(data2['host_response_rate'], errors='coerce')
data2['host_acceptance_rate'] = pd.to_numeric(data2['host_acceptance_rate'], errors='coerce')
data2['host_is_superhost'] = pd.to_numeric(data2['host_is_superhost'], errors='coerce')
data2['host_has_profile_pic'] = pd.to_numeric(data2['host_has_profile_pic'], errors='coerce')
data2['host_identity_verified'] = pd.to_numeric(data2['host_identity_verified'], errors='coerce')
data2['instant_bookable'] = pd.to_numeric(data2['instant_bookable'], errors='coerce')
data2['price'] = data2['price'].str.replace(r'[\$,]', '', regex=True).astype(float)
```

```python
# Se extrae el año de la última review y la del año donde inició el host de cada lugar
data2["last_review"]
data2["host_since"]
# Usamos una expresión regular para extraer el año
data2['last_review'] = data2['last_review'].str.extract(r'(\d{4})')
data2['last_review'] = pd.to_numeric(data2['last_review'], errors='coerce')
data2['host_since'] = data2['host_since'].str.extract(r'(\d{4})')
data2['host_since'] = pd.to_numeric(data2['host_since'], errors='coerce')
# Se agrega una nueva columna
data2['host_years'] = data2["last_review"].max() - data2['host_since']
```

```python
# Borramos las cualitativas
data2.drop(["description","neighborhood_overview","host_about","host_neighbourhood","name","neighbourhood","host_location","first_review","has_availability"], axis = 1, inplace = True)
# Borramos las cuantitativas
data2.drop(["scrape_id","minimum_nights","maximum_nights","minimum_minimum_nights","minimum_maximum_nights", "maximum_minimum_nights", "maximum_maximum_nights","availability_60", "availability_90", "calculated_host_listings_count_entire_homes", "calculated_host_listings_count_private_rooms", "calculated_host_listings_count_shared_rooms","last_review"], axis = 1, inplace = True)
```

```python
# Dividimos por valores cuantitativos y cualitativos
cuantis = data2.select_dtypes(np.number)
cualis = data2.select_dtypes(include=['object'])
```

## Sustitución de valores nulos

```python
# Identificamos valores cualitativos nulos
cualis.isnull().sum()
```

```python
#Creamos una copia para manipulación
cualis_limpia = cualis.copy()
```

```python
# Se sutituyen los valores nulos por la moda dadas las categorías y distribución de los datos
moda = cualis_limpia["host_response_time"].value_counts().index[0]
cualis_limpia["host_response_time"] = cualis_limpia["host_response_time"].fillna(moda)
```

```python
cualis_limpia["host_response_time"].isnull().sum()
```

```python
# Al contar con solo 25 datos nulos de igual manera se sustituirá con la moda
moda1 = cualis_limpia["bathrooms_text"].value_counts().index[0]
cualis_limpia["bathrooms_text"] = cualis_limpia["bathrooms_text"].fillna(moda)
```

```python
# Corroboramos que no tengamos valores nulos
cualis_limpia.isnull().sum()
```

```python
# Generamos una copia para la limpieza de cuantitativos
cuantis_limpia = cuantis.copy()
```

```python
# Identificar valores cualitativos nulos
cuantis_limpia.isnull().sum()
```

```python
cuantis[["host_response_rate","host_acceptance_rate"]].head(5)
# Dado que presentan porcentajes se sustituye valores nulos con la media
cuantis_limpia["host_response_rate"] = cuantis_limpia["host_response_rate"].fillna(round(cuantis["host_response_rate"].mean(),1))
cuantis_limpia["host_acceptance_rate"] = cuantis_limpia["host_acceptance_rate"].fillna(round(cuantis["host_acceptance_rate"].mean(),1))
```

```python
cuantis["host_is_superhost"].value_counts()
# Reemplazaremos por un nuevo valor
cuantis_limpia["host_is_superhost"] = cuantis_limpia["host_is_superhost"].fillna(2)
```

```python
# Para los baños, habitaciones y camas se sustituirán valores nulos por la mediana para no afectar la distribución
cuantis_limpia["bathrooms"] = cuantis_limpia["bathrooms"].fillna(int(round(cuantis["bathrooms"].median(), 0)))
cuantis_limpia["bedrooms"] = cuantis_limpia["bedrooms"].fillna(int(round(cuantis["bedrooms"].median(), 0)))
cuantis_limpia["beds"] = cuantis_limpia["beds"].fillna(int(round(cuantis["beds"].median(), 0)))
```

```python
# Para la columna de "precio" se sustituirán los valores nulos por la mediana para no afectar la distribución
cuantis_limpia["price"] = cuantis_limpia["price"].fillna(round(cuantis["price"].median(), 0))
```

```python
# Columnas que inician con 'review_scores', los valores son escala Likert
dfReviewScores = cuantis_limpia.loc[:, cuantis_limpia.columns.str.contains("review_scores")]
[dfReviewScores.min(), dfReviewScores.max()]
```

```python
# Se sustituyen los valores núlos de escala Likert con la media
cuantis_limpia["review_scores_rating"] = cuantis_limpia["review_scores_rating"].fillna(round(cuantis["review_scores_rating"].mean(), 2))
cuantis_limpia["review_scores_accuracy"] = cuantis_limpia["review_scores_accuracy"].fillna(round(cuantis["review_scores_accuracy"].mean(), 2))
cuantis_limpia["review_scores_cleanliness"] = cuantis_limpia["review_scores_cleanliness"].fillna(round(cuantis["review_scores_cleanliness"].mean(), 2))
cuantis_limpia["review_scores_checkin"] = cuantis_limpia["review_scores_checkin"].fillna(round(cuantis["review_scores_checkin"].mean(), 2))
cuantis_limpia["review_scores_communication"] = cuantis_limpia["review_scores_communication"].fillna(round(cuantis["review_scores_communication"].mean(), 2))
cuantis_limpia["review_scores_location"] = cuantis_limpia["review_scores_location"].fillna(round(cuantis["review_scores_location"].mean(), 2))
cuantis_limpia["review_scores_value"] = cuantis_limpia["review_scores_value"].fillna(round(cuantis["review_scores_value"].mean(), 2))
```

```python
# Se redondean a valor entero

cuantis_limpia["review_scores_rating"] = cuantis_limpia["review_scores_rating"].round()
cuantis_limpia["review_scores_accuracy"] = cuantis_limpia["review_scores_accuracy"].round()
cuantis_limpia["review_scores_cleanliness"] = cuantis_limpia["review_scores_cleanliness"].round()
cuantis_limpia["review_scores_checkin"] = cuantis_limpia["review_scores_checkin"].round()
cuantis_limpia["review_scores_communication"] = cuantis_limpia["review_scores_communication"].round()
cuantis_limpia["review_scores_location"] = cuantis_limpia["review_scores_location"].round()
cuantis_limpia["review_scores_value"] = cuantis_limpia["review_scores_value"].round()
```

```python
# Se sustituyen los valores nulos de reviews_per_month por la mediana
cuantis_limpia["reviews_per_month"] = cuantis_limpia["reviews_per_month"].fillna(round(cuantis["reviews_per_month"].median(), 2))
```

```python
# Corroboramos que no tengamos valores nulos
cuantis_limpia.isnull().sum().sum()
```

```python
# Unimos ambas limpiezas (Valores cuantitativos y cualitativos limpios)
dfMexicoLimpio = pd.concat([cualis_limpia, cuantis_limpia], axis=1)
```
